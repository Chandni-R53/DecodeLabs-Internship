### Data Cleaning

### Identifying Missing Values

```python
df.isnull().sum()
```

### Handling Missing Values in CouponCode

```python
df['CouponCode']=df['CouponCode'].fillna('NO COUPON')
```

```python
df.isnull().sum()
```

### Check for Duplicate Values

```python
print(df.duplicated().sum())
```

```python
print(df['CustomerID'].duplicated().sum())
```

### Observation

The OrderID column has no duplicates it uniquely identifies 1200 rows whereas the CustomerID column contains 11 repeated values, it can be possible because a single customer can place multiple orders.

### Data Validation

```python
df['Quantity'].describe()
```

```python
df['UnitPrice'].describe()
```

```python
df['CalculatedPrice']=df['Quantity']*df['UnitPrice']
```

```python
print((df['CalculatedPrice']!=df['TotalPrice']).sum())
```

```python
print((round(df['CalculatedPrice'],2)!=round(df['TotalPrice'],2)).sum())
```

```python
df=df.drop(columns=['CalculatedPrice'])
```

```python
df.shape
```

### Total Price Validation

The TotalPrice column was validated by recalculating order values using Quantity and UnitPrice.

### Data Preparation

### Feature Creation from Date Column

Additional date-based features were created from the Date column, including Year and Month for future analysis.

```python
df['Year']=df['Date'].dt.year
```

```python
df['Month']=df['Date'].dt.month
```
